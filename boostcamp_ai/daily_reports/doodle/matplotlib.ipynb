```python
import random
import numpy as np
x1 = range(100)
y1 = [np.cos(value) for value in x1]

x2 = range(100)
y2 = [np.sin(value) for value in x2]
```

```python
plt.plot(x2, y2)
```

```python
n = 50
x = np.random.rand(n)
y = np.random.rand(n)
colors = np.random.rand(n)
area = np.pi * (15 * np.random.rand(n))**2
plt.scatter(
    x, y, 
    s=area, # 데이터 크기 지정 
    c=colors, 
    alpha=.5 # # 투명도
)
plt.show()
```

```python
data = [[5, 25, 50, 20],
[4, 23, 51, 17],
[6, 22, 52, 19]]

x = np.arange(4)
plt.bar(x, data[0], color='b', width=.25)
plt.bar(x + 0.25, data[1], color='c', width=.25)
plt.bar(x + 0.5, data[2], color='g', width=.25)
plt.xticks(x+.25, ['A','B','C','D'])
plt.show()
```
